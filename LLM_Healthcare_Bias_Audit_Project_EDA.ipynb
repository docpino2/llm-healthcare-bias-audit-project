```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv("/content/base_cases.csv")
demo = pd.read_csv("/content/demographics.csv")
inf = pd.read_csv("/content/inferences_skeleton.csv")
```

## Specialty distribution

```python
base['specialty'].value_counts().plot(kind='bar')
plt.title('Distribution of Clinical Specialties')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# ---------------------------------------------------
# Clinical Scenario Distribution (Base Cases)
# ---------------------------------------------------

plt.figure(figsize=(10,6))

(
    base["case_title"]
    .value_counts()
    .sort_values(ascending=False)
    .plot(kind="bar")
)

plt.title("Distribution of Clinical Scenarios (Base Cases)")
plt.xlabel("Clinical Scenario")
plt.ylabel("Count")
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()

print("Total unique clinical scenarios:", base["case_title"].nunique())
```

## Acuity distribution

```python
base['acuity_level'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribution of Acuity Levels')
plt.show()
```

## Demographic balance

```python
pivot = demo.pivot_table(index='gender', columns='ethnicity', values='demo_id', aggfunc='count')
plt.imshow(pivot.values)
plt.colorbar()
plt.xticks(range(len(pivot.columns)), pivot.columns, rotation=45)
plt.yticks(range(len(pivot.index)), pivot.index)
plt.show()
```

## Inferences per model

```python
inf['model_name'].value_counts().plot(kind='bar')
plt.title('Number of Inferences per Model')
plt.show()
```

## Dataset size summary

```python
print('Base cases:', len(base))
print('Demographics:', len(demo))
print('Models:', inf['model_name'].nunique())
print('Total inferences:', len(inf))
```
